```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
 
I = cv.imread("C:/Users/Victor Manuel/Documents/John/hand.jpg")
 
```

```python
img = cv.cvtColor(I, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
first_layer=cv.pyrDown(img)
second_layer=cv.pyrDown(first_layer)
third_layer=cv.pyrDown(second_layer)
fourth_layer=cv.pyrDown(third_layer)
plt.figure(figsize=(20,25)) #DIMENSIONA LAS IMAGENES

plt.subplot(151),plt.imshow(img)
plt.subplot(152),plt.imshow(first_layer)
plt.subplot(153),plt.imshow(second_layer)
plt.subplot(154),plt.imshow(third_layer)
plt.subplot(155),plt.imshow(fourth_layer)
```

```python
expanded_layer=cv.pyrUp(first_layer)
expanded_layer2=cv.pyrUp(second_layer)
expanded_layer3=cv.pyrUp(third_layer)
expanded_layer4=cv.pyrUp(fourth_layer)

plt.figure(figsize=(20,25)) #DIMENSIONA LAS IMAGENES

plt.subplot(151),plt.imshow(expanded_layer)
plt.subplot(152),plt.imshow(expanded_layer2)
plt.subplot(153),plt.imshow(expanded_layer3)
plt.subplot(154),plt.imshow(expanded_layer4)
```

```python
laplacian_layer=cv.subtract (img,expanded_layer)
laplacian_layer2=cv.subtract (first_layer,expanded_layer2)
plt.figure(figsize=(20,25)) #DIMENSIONA LAS IMAGENES
plt.subplot(151),plt.imshow(laplacian_layer)
plt.subplot(152),plt.imshow(laplacian_layer2)


```
